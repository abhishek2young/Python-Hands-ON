# 7. How will ypu explain reindexing in Pandas?

* In Pandas by using set_index() method we can do the reindexing.
> Changing the index >>
* When you reindex a DataFrame or Series,you specify a new set of labels for the rows and columns.
* These new labels can be a modified version of the existing labels a completely new set of labels.
> Aligning data 
* The primary benefit of reindexing is data alignment.
* WHen you perform operations on multiple data structures dataframes together Pandas aligns the data based on the index labels.
* Missing labels are filled with Nan values ensuring that data is combined correctly.

# 27. What is the use of apply function in Pandas?

* We can apply any function on dataframe directly , any functions means built in user define , lambda , numpy function etc.
* The apply() function in pandas is a powerful and flexible method that is used in a given function along with an axis of a dataframe or series.
* It allows you to perform complex operations on data by applying your own custom or even built-in function , to individual elements , rows or columns of the data.

```python
arr1 = np.random.randint(1,20,size=(4,2))
df = pd.DataFrame(arr1)
df.apply(np.sqrt)
```

```python
# Apply function in pandas is use to apply function on whole data frame
df = pd.DataFrame(np.random.randint(1,9,size=(3,4)))
df
```

```python
df.apply(lambda x : x**2)
```

```python
def cube(x):
    val = x**3
    return val
df.apply(cube)
```

# 31. What is the map function in Pandas?

```python
"""The Map function in Pandas is used to substitute each value in Series or dataframe column with 
another value.It is a versatile and convenient way to perform element-wise transformations on 
data."""
d1 = {"Employee":["Saniya","Sara","Alexa","Jenny"]}
df = pd.DataFrame(d1)
d2 = {"Saniya": "Software Developer" , "Sara":"Developer" , "Alexa":"Assistant"}
df["designation"] = df["Employee"].map(d2)
df
```

# 35. How to get the items which are not common to both series A and Series B?

```python
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([4,5,6,7,8])
s11 = set(s1)
s22 = set(s2)
result = s11.symmetric_difference(s22)
final = pd.Series(list(result))
final
```

# 36. How to get the minimum , 25th percentile , median , 75th percentile and max of a numeric series

```python
s1 = pd.Series([1,2,3,4,5])
print("Median:",s1.median())
print("Minimum:",s1.min())
print("25th percentile:",s1.quantile(0.25))
print("75th percentile:",s1.quantile(0.75))
print("Maximum:",s1.max())
```
