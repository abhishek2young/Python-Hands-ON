# 26. Write a pandas program to extract years between 1800 and 2200 from the specified column of a given dataframe

```python
import pandas as pd

# Assuming you have a DataFrame named 'df' and a column named 'year'
data = {'year': [1700, 1850, 2000, 2300, 1950]}
df = pd.DataFrame(data)

# Extract years between 1800 and 2200
filtered_df = df[(df['year'] >= 1800) & (df['year'] <= 2200)]

# Print the result
print("DataFrame with years between 1800 and 2200:")
print(filtered_df)
```

# 33. Write a pandas program to merge two given datasets using multiple join keys
